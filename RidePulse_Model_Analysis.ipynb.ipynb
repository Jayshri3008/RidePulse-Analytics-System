```python
def predict_price(cab_name, source, surge_multiplier, icon):

    import numpy as np
    import pandas as pd

    # full zero array (same size as training data)
    x = np.zeros(len(X.columns))

    # manually set important features (index adjust if needed)
    x[0] = cab_name
    x[1] = source
    x[2] = surge_multiplier
    x[3] = icon

    # convert to dataframe with correct column names
    input_data = pd.DataFrame([x], columns=X.columns)

    return random.predict(input_data)[0]
```
